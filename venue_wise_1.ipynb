```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('each_ball_records.csv')
```

```python
df['batsman_score'] = pd.to_numeric(df.outcome, errors='coerce')
df['1s'] = (df['outcome'] == '1').astype('int')
df['2s'] = (df['outcome'] == '2').astype('int')
df['3s'] = (df['outcome'] == '3').astype('int')
df['4s'] = (df['outcome'] == '4').astype('int')
df['6s'] = (df['outcome'] == '6').astype('int')
df['w'] = (df['outcome'] == 'w').astype('int')
```

```python
def wide(y):
    return y.str.contains('wd')
def noball(z):
    return z.str.contains('nb')
```

```python
df['wd'] = wide(df['outcome']).astype('int')
df['nb'] = noball(df['outcome']).astype('int')
```

```python
merged_df['Phase'] = merged_df['over_no'].apply(phase)
merged_df.tail()
```

```python
venue = merged_df[['venue','location','match_no','inningno','score','over_no','Phase','w']]
venue
```

```python
venue_wise = venue.pivot_table(venue, index=['venue','location','match_no','inningno','over_no','Phase'], aggfunc=sum)
temp = venue_wise.reset_index()
temp
```

```python
nms_phase = temp.pivot_table(values=['score','w'], index = ['location','venue'], columns =['inningno','Phase'], aggfunc=sum)
nms_phase
```

```python
nms_calc = nms_phase.query('venue == "Narendra Modi Stadium"')
nms_calc
```

```python
nms_avg = nms_calc / 9
nms_avg
```

```python
stacked_nms = nms_avg.stack(level=['inningno', 'Phase'])
stacked_nms
```

```python
nms_pivoted = stacked_nms.reset_index()
nms_pivoted
```

```python
nms_final = nms_pivoted.drop(columns=['location','venue'],axis=0)
nms_final
```
